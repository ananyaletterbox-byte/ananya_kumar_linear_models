```python
#ChatGPT was used as a guide for this assignment
#In collaboration with Jaitha Jasti
'''
QUESTION 1:
1.1 A model is linear in regards to its paramters/coefficients where they appear in a linear way (using weighted sums); however, the variables are not necessarily linear and if the coefficients are altered, so will the linearity of the model.
1.2 The coefficient portrays the difference from the baseline when all other variables are constant. The intercept is what corresponds to the baseline group and is when all dummies = 0. So each dummy coefficient describes how much that category differs from the baseline.
1.3 While linear regression can be used to predict continuous values, it does not predict probabilities (which remains between 0 and 1). It can tell you if predictions are less than 0 or greater than 1 and assumes that variance is constant and errors are normally distributed, which means that classification would not be feasible.
1.4 If you have high training accuracy but poor test/validation performance, then the linear model is overfitting. Also, if the model captures a lot of noise and if adding more variables improves training fit but makes generalization worse, then the mopdel is overfitting. If coefficients change a lot when you have even the smallest of data changes, then that is also an indication of overfitting. 
1.5 Multicollinearity occurs when your predictors are highly correlated, which makes it difficult to separate the effects of each of them individually. Using 2SLS, stage 1 would have the regression of one variable (x1) on other predictors to get the predicted values of xhat1. Then stage 2 would use those predicted values, which is not less correlated with the other variables, in your primary/main regression. By doing so, you isolate the variation that is coming from x1 to reduce the instability that is caused by multicollinearity. 
1.6 You can incorporate nonlinear relationships by transforming your inputs (features/control/response/independent variables) via polynomial, interaction, log/exponential, or piecewise functions (for example). While the model is still linear since it is linear in its coefficients, the variables are technically nonlinear and from there you can analyze your target/response/dependent/outcome variable.
1.7 The intercept is the predicted value of y when all predictors are equal to 0 (for dummies, this represents the baseline category). The slope is the change in y for every one unit increase in xi, while all the other variables are held constant. Finally, the dummy variable coefficient is the difference in y between that specific category and the reference category, while holding other variables constant.
'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#loading data + cleaning names
df = pd.read_csv("/Users/ananyakumar/Downloads/3001/Q1_clean.csv")


# Clean column names (remove spaces)
df = df.rename(columns={
 "Review Scores Rating": "ReviewScore",
 "Neighbourhood ": "Neighbourhood",
 "Property Type": "PropertyType",
 "Room Type": "RoomType",
 "Price per night": "Price"
})

#average price and score by neighborhood
grouped = df.groupby("Neighbourhood")[["Price", "ReviewScore"]].mean()
print(grouped)

#ANSWER 2.1 The most expensive bourough on average is Manhattan.

#create kernel density plots
# Price KDE
sns.kdeplot(data=df, x="Price", hue="Neighbourhood")
plt.title("KDE of Price")
plt.show()

# Log price KDE
df["log_price"] = np.log(df["Price"])

sns.kdeplot(data=df, x="log_price", hue="Neighbourhood")
plt.title("KDE of Log Price")
plt.show()

#regression without intercept (all dummies)

dummies = pd.get_dummies(df["Neighbourhood"])

X = dummies
y = df["Price"]

model = sm.OLS(y, X).fit()
print(model.summary())
#ANSWER 2.2: The regression coefficients exactly match the average prices from part 1. This shows that the regression of a continuous variable on one categorical variable produces coefficients that are ecactly equal to the category means. So if you conlude a constant and leave just one category out (reference)m then the coefficents will represent differnces from the reference rather than raw means. Here, the R squared is low (0.046) which means that Neighborhood explains only around 4.6% of the variation in price (price varies a lot within neighborhoods)

#regression with intercept:

dummies = pd.get_dummies(df['Neighbourhood'], drop_first=True) # Bronx dropped
X = sm.add_constant(dummies) # constant = mean price for Bronx

# Remove $ or commas if they exist, then convert to float
df['Price'] = df['Price'].replace('[\$,]', '', regex=True).astype(float)
df = df.dropna(subset=['Price', 'Neighbourhood'])
dummies = dummies.astype(float)
import statsmodels.api as sm
X = sm.add_constant(dummies)
y = df['Price'].astype(float)
model2 = sm.OLS(y, X).fit()
print(model2.summary())

#ANSWER 2.3: I had to handle the creation of dummies differently by dropping Bronx as the reference category; you must always drop one category when including a constant, otherwise you will get perfect multicollinearity (where sum of dummies = 1) and the model will fail to invert the matrix; the dropped category (Bronx) now became my baseline and all the other coefficients are then interpreted relative to that baseline. 
#The intercept is $75.28, which represents the expected Price for the reference category (Bronx) because all the other dummy variables are equal to 0 for Bronx; this means if the property is in Bronx, then the predicted price is $75.28. 
#The interpretations for the Brooklyn coefficient (52.47) is that Brooklyn properties are expected to be $52.47 higher than Bronx, as long as everything else is equal. The Manhattan coefficent is 108.39, which means the Manhattan properties are expected to be $108.39 higher than Bronx. The Queens coefficient is 21.58, which means that the Queens properties are expected to be $21.58 higher than Bronx. Finally, the Staten Island coefficient is 70.89, which means that the Staten Island properties are expected to be $70.89 higher than Bronx.
# All in all, if it is positive, then it is more expensive than the Bronx and if it is negative it is cheaper but here none of them are cheaper. 
#You can get the coefficients in part 2 from these new coefficients in that if the intercept is the Bronx mean, then the coeff(Manhattan) is +100 so Manhattan mean is intercept + 100 which means that the full mean is equal to the baseline + dummy coefficient. 
#train/test split + regression 

df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_test = train_test_split(df, test_size=0.2, random_state=42)

# Neighbourhood dummies
train_dummies = pd.get_dummies(df_train["Neighbourhood"], drop_first=True).astype(float)
test_dummies = pd.get_dummies(df_test["Neighbourhood"], drop_first=True).astype(float)

train_dummies, test_dummies = train_dummies.align(test_dummies, join="left", axis=1, fill_value=0)

# Numeric feature
train_num = df_train[["ReviewScore"]].apply(pd.to_numeric, errors="coerce")
test_num = df_test[["ReviewScore"]].apply(pd.to_numeric, errors="coerce")

# Clean Price
df_train["Price"] = pd.to_numeric(df_train["Price"].replace(r"[\$,]", "", regex=True), errors="coerce")
df_test["Price"] = pd.to_numeric(df_test["Price"].replace(r"[\$,]", "", regex=True), errors="coerce")

# Combine features
X_train = pd.concat([train_num, train_dummies], axis=1)
X_test = pd.concat([test_num, test_dummies], axis=1)

# Clean bad values
X_train = X_train.replace([np.inf, -np.inf], np.nan).dropna()
X_test = X_test.replace([np.inf, -np.inf], np.nan).dropna()

# Align y with cleaned X
y_train = df_train["Price"].loc[X_train.index].astype(float)
y_test = df_test["Price"].loc[X_test.index].astype(float)

# Add intercept
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Fit OLS
model3 = sm.OLS(y_train.astype(float), X_train.astype(float)).fit()

# Predict
y_pred = model3.predict(X_test.astype(float))

#from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import mean_squared_error, r2_score
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE:", rmse)
print("Test R^2:", r2)
print(model3.summary())

# Dummies
train_dummies = pd.get_dummies(df_train["Neighbourhood"], drop_first=True).astype(float)
test_dummies = pd.get_dummies(df_test["Neighbourhood"], drop_first=True).astype(float)

# Align test dummies to train columns
test_dummies = test_dummies.reindex(columns=train_dummies.columns, fill_value=0)

# Combine dummies with other features
X_train = pd.concat([train_dummies, df_train["ReviewScore"].astype(float)], axis=1)
X_test = pd.concat([test_dummies, df_test["ReviewScore"].astype(float)], axis=1)

# Add constant for intercept
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Target
y_train = df_train["Price"].astype(float)
y_test = df_test["Price"].astype(float)

# Fit OLS model
model3 = sm.OLS(y_train, X_train).fit()

# Predictions
y_pred = model3.predict(X_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = model3.rsquared

print("RMSE:", rmse)
print("R^2:", r2)
print(model3.summary())

#ANSWER 2.4: The R squared is 0.05 and the RMSE is 140.92. The coefficient on Review Score Rating is 1.2119. This means that for each 1-point increase in the review score, the expected price increases by 1.2119 units (if Price is in the thousands, then this is a $1.21 increase); this is significantly significant because p < 0.001 (0.000). 
#The most expensive kind of property is determined by that the constant represents the baseline category, which is Bronx where the coefficients for neighborhoods are additive adjustments relative to that reference.
#This is the breakdown:

'''
Constant = -35.1573 (baseline: Bronx)
Brooklyn = 51.5560 → Predicted price = -35.1573 + 51.5560 ≈ 16.4
Manhattan = 107.4136 → Predicted price = -35.1573 + 107.4136 ≈ 72.26
Queens = 20.1986 → Predicted price ≈ -14.96
Staten Island = 23.4547 → Predicted price ≈ -11.7
''' 
#Thus, Manhattan has the highest predictive price, which means that the most expensive type of property that you can rent is in Manhattan. (adding ReviewScore might shift each up)

train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
# Create dummies first
train_prop = pd.get_dummies(train_df["PropertyType"], drop_first=True)
test_prop = pd.get_dummies(test_df["PropertyType"], drop_first=True)

# Align test dummies to train columns (already done)
test_prop = test_prop.reindex(columns=train_prop.columns, fill_value=0)

# Combine features
X_train2 = pd.concat([train_dummies, train_prop, df_train["ReviewScore"].astype(float)], axis=1)
X_test2 = pd.concat([test_dummies, test_prop, df_test["ReviewScore"].astype(float)], axis=1)

# Add constant
X_train2 = sm.add_constant(X_train2)
X_test2 = sm.add_constant(X_test2)

# Ensure target is float
y_train = df_train["Price"].astype(float)
y_test = df_test["Price"].astype(float)

# Fit OLS
model4 = sm.OLS(y_train.astype(float), X_train2.astype(float)).fit()

# Predictions
y_pred2 = model4.predict(X_test2)

# Metrics
rmse2 = np.sqrt(mean_squared_error(y_test, y_pred2))
r2_2 = model4.rsquared

print("RMSE:", rmse2)
print("R^2:", r2_2)
print(model4.summary())

#ANSWER 2.5: The R squared is 0.06. The RMSE is 140.3023. The coefficient on Review Score is 1.201. This means that for each 1-point increase in ReviewScore, the predicted Price increases by around $1.20, as long as Neighborhood and PropertyType remains constant. 
#Looking at the coefficients of PropertyType, the highest positive coefficient is Bungalow = 83.97. Then, if you look at Neighborhoods, which we determine that Manhattan is the most expensive, this means that the most expensive type of property you can rent is likely a Bungalow in Manhattan. 

#ANSWER 2.6: In part 4, the model only included Neighborhood dummies and ReviewScore where the coefficient on ReviewScore is around 1.21. This means that if Neighborhood is constant, then each additional point in ReviewScore increases the predicted Price by around $1.21
#In part 5, we added PropertyType dummies which made the coefficient on ReviewScore drop to $1.20. So this means that if you hold both Neighborhood and Property Type constant, then each additional point in ReviewScore increases the predicted Price by around $1.20.
#The coefficient changed because in multiple linear regression, the coefficient for a given variable (which here is ReviewScore) represents only its partial effect (the effect of ReviewScore after controlling for all the other variables involved in the model). So when you add PropertyType, some of the variation in Price that was previously explained by ReviewScore may not also be explained by PropertyType which can make the ReviewScore coefficent smaller or larger, depending on the correlations between ReviewScore and the new vars. Coefficients can change!

#ANSWER 2.7: N/A (optional)



#QUESTION 3:

#loading data:
df = pd.read_csv("/Users/ananyakumar/Downloads/3001/cars_hw.csv")
# Summary statistics for Price
print(df['Price'].describe())

# Kernel density plot
sns.kdeplot(df['Price'], fill=True)
plt.title("Kernel Density of Car Prices")
plt.xlabel("Price")
plt.ylabel("Density")
plt.show()

#summary by car brand
# Group by Make and summarize
make_summary = df.groupby("Make")["Price"].describe()
print(make_summary)

# Optional: top 10 most expensive brands by median (Chat)
print(make_summary['50%'].sort_values(ascending=False).head(10))

# Grouped KDE plot
for make in df['Make'].unique():
 sns.kdeplot(df[df['Make']==make]['Price'], label=make, alpha=0.6)

plt.title("Kernel Density of Prices by Make")
plt.xlabel("Price")
plt.ylabel("Density")
plt.legend()
plt.show()

#ANSWER 3.2: The most expensive car brands (the ones with the highest price distributions) are Jeep and MG Motors (these are the ones with density curves furthest to the right). Prices are gnerally clustered between 0 and 2 million units where the majority of brands have price distributions around 0.5 to 1.5 on the price axis. There is a very high density around 0.5 mark, which tells us that many vehicles fall within this lower price range. Overall, the distributions for most car brands are narrow and clustered together, which means there is probably a pretty common market price range for most of the brands.

#split into training and test sets
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Extract target
y_train = train_df['Price'].astype(float)
y_test = test_df['Price'].astype(float)



#model 1 (numeric vars only)

# Select numeric columns
numeric_cols = ['Seating_Capacity', 'Mileage_Run', 'Make_Year'] # Adjust based on your dataset

X_train_num = train_df[numeric_cols].astype(float)
X_test_num = test_df[numeric_cols].astype(float)

# Add constant
X_train_num = sm.add_constant(X_train_num)
X_test_num = sm.add_constant(X_test_num)

# Fit OLS
model_num = sm.OLS(y_train, X_train_num).fit()

# Predictions and metrics
y_pred_train = model_num.predict(X_train_num)
y_pred_test = model_num.predict(X_test_num)

rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_train = model_num.rsquared
from sklearn.metrics import r2_score, mean_squared_error
r2_test = r2_score(y_test, y_pred_test)

print("Numeric Model")
print("R^2 (train):", r2_train)
print("R^2 (test):", r2_test)
print("RMSE (train):", rmse_train)
print("RMSE (test):", rmse_test)

#ANSWER 3.4a: R squared of the training set is 0.3144. R squared of the test set is 0.2941. The RMSE of the training set is 309338.37. The RMSE of the test set is 284546.95.

#model 2 (categorical vars only)
# One-hot encode categorical variables
categorical_cols = ['Color', 'Body_Type', 'Make', 'Fuel_Type', 'Transmission', 'Transmission_Type']
X_train_cat = pd.get_dummies(train_df[categorical_cols], drop_first=True).astype(float)
X_test_cat = pd.get_dummies(test_df[categorical_cols], drop_first=True).astype(float)

# Align test to train
X_test_cat = X_test_cat.reindex(columns=X_train_cat.columns, fill_value=0)

# Add constant
X_train_cat = sm.add_constant(X_train_cat)
X_test_cat = sm.add_constant(X_test_cat)

# Fit model
model_cat = sm.OLS(y_train, X_train_cat).fit()

# Predictions and metrics
y_pred_test_cat = model_cat.predict(X_test_cat)
rmse_test_cat = np.sqrt(mean_squared_error(y_test, y_pred_test_cat))
r2_test_cat = r2_score(y_test, y_pred_test_cat)

print("Categorical Model")
print("R^2 (test):", r2_test_cat)
print("RMSE (test):", rmse_test_cat)

#ANSWER 3.4b: The R squared of the test set is 0.7081. The RSME of the test set is 182974.738.
#Model 2 performs much better on the test set, which is indicated by the higher R squared and the lower RMSE. This means that the categorical variables like Make, Body_Type, Transmission explain a large portion of the variation in car prices compared to solely the numeric variables.



#third model (combining numeric and categorical features)

# Combine features
X_train_combined = pd.concat(
    [train_df[numeric_cols], X_train_cat.drop(columns="const")],
    axis=1
)

X_test_combined = pd.concat(
    [test_df[numeric_cols], X_test_cat.drop(columns="const")],
    axis=1
)

# Convert to numeric safely
X_train_combined = X_train_combined.apply(pd.to_numeric, errors="coerce")
X_test_combined = X_test_combined.apply(pd.to_numeric, errors="coerce")

# Drop rows with NaNs (IMPORTANT)
X_train_combined = X_train_combined.dropna()
X_test_combined = X_test_combined.dropna()

# ALIGN y with X (THIS FIXES YOUR ERROR)
y_train_aligned = train_df.loc[X_train_combined.index, "Price"].astype(float)
y_test_aligned = test_df.loc[X_test_combined.index, "Price"].astype(float)

# Add constant
X_train_combined = sm.add_constant(X_train_combined, has_constant="add")
X_test_combined = sm.add_constant(X_test_combined, has_constant="add")

# Fit model
model_combined = sm.OLS(y_train_aligned, X_train_combined).fit()

# Predict
y_pred_combined = model_combined.predict(X_test_combined)

# Metrics
rmse_combined = np.sqrt(mean_squared_error(y_test_aligned, y_pred_combined))
r2_combined = r2_score(y_test_aligned, y_pred_combined)

print("Joint Model")
print("R^2 (test):", r2_combined)
print("RMSE (test):", rmse_combined)

#ANSWER 3.4c: The R squared on the test set is 0.822. The RSME on the test set is 142829.739
#The joint model worked the best. It did better than the catgorical's (previous best) R-squared by 0.1139 so joint explains around 11.4% more variance. RMSE improvement showed a prediction error decrease by 40,145 units. 


#poly work
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import pandas as pd
import numpy as np

# Make sure these exist:
# train_df, test_df, numeric_cols, y_train, y_test

degrees = [1, 2, 3, 4]
results = []

for deg in degrees:
    # 1. Build clean numeric X for this loop
    X_train_num = train_df[numeric_cols].apply(pd.to_numeric, errors="coerce")
    X_test_num = test_df[numeric_cols].apply(pd.to_numeric, errors="coerce")
    X_train_num = X_train_num.replace([np.inf, -np.inf], np.nan).dropna()
    X_test_num = X_test_num.replace([np.inf, -np.inf], np.nan).dropna()
    #2. Align y to the cleaned X rows
    y_train_loop = train_df.loc[X_train_num.index, "Price"].astype(float)
    y_test_loop = test_df.loc[X_test_num.index, "Price"].astype(float)
    # 3. Create polynomial features
    poly = PolynomialFeatures(degree=deg, include_bias=False)
    X_train_poly = poly.fit_transform(X_train_num)
    X_test_poly = poly.transform(X_test_num)
    # 4. Convert to DataFrame with the CORRECT index
    X_train_poly = pd.DataFrame(
        X_train_poly,
        columns=poly.get_feature_names_out(numeric_cols),
        index=X_train_num.index
    )
    X_test_poly = pd.DataFrame(
        X_test_poly,
        columns=poly.get_feature_names_out(numeric_cols),
        index=X_test_num.index
    )
    # 5. Add constant for intercept
    X_train_poly = sm.add_constant(X_train_poly, has_constant="add")
    X_test_poly = sm.add_constant(X_test_poly, has_constant="add")
    # 6. Fit OLS model
    model = sm.OLS(y_train_loop, X_train_poly).fit()
    # 7. Predict
    y_pred_train = model.predict(X_train_poly)
    y_pred_test = model.predict(X_test_poly)
    # 8. Compute metrics
    rmse_train = np.sqrt(mean_squared_error(y_train_loop, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test_loop, y_pred_test))
    r2_train = r2_score(y_train_loop, y_pred_train)
    r2_test = r2_score(y_test_loop, y_pred_test)
    # 9. Store results
    results.append({
        "degree": deg,
        "rmse_train": rmse_train,
        "rmse_test": rmse_test,
        "r2_train": r2_train,
        "r2_test": r2_test
    })
    # 10. Print summary
    print(f"Degree {deg}: RMSE_train = {rmse_train:.2f}, "
        f"RMSE_test = {rmse_test:.2f}, "
    f"R²_train = {r2_train:.3f}, R²_test = {r2_test:.3f}")

#ANSWER 3.5:

'''
Degree 1: RMSE_train = 309338.37, RMSE_test = 284546.95, R²_train = 0.314, R²_test = 0.294
Degree 2: RMSE_train = 304067.02, RMSE_test = 278385.20, R²_train = 0.338, R²_test = 0.324
Degree 3: RMSE_train = 302028.33, RMSE_test = 277525.40, R²_train = 0.346, R²_test = 0.328
Degree 4: RMSE_train = 301021.00, RMSE_test = 278391.91, R²_train = 0.351, R²_test = 0.324
'''
#As the degree increases, the RMSE_train decreases. The RMSE_test also decreases at the start but then increases slightly at degree 4.
#This means that the training error keeps improving with more complexity (higher-degree polynomial) but the test error stops improving and then slightly increases at degree 4 which could indicate overfitting.
#R squared_train increases with degree and R squared_test increases initially and then decreases at degree 4. This means that the model fits the training data better with higher-degree terms but the generalization stops improving at degree 4. So, if you keep increasing the degree, R squared_test could become negative which means the model is worse than when you just predict the mean of the target. 
#The best test R squared occurs at degree 3. Part 4's best model had a much better model with a higher R sqaured for the test set which tells us that more features/higher degree does not automatically improve R squared test.



# using best model (joint)

# Predict using the joint model
y_pred_test = model_combined.predict(X_test_combined)

# True values
y_true = y_test_aligned  # Make sure indices match

# Scatter plot: predicted vs. true
plt.figure(figsize=(7,7))
plt.scatter(y_true, y_pred_test, alpha=0.6)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')  # diagonal
plt.xlabel("True Prices")
plt.ylabel("Predicted Prices")
plt.title("Predicted vs True Prices (Test Set)")
plt.show()

# Residuals
residuals = y_true - y_pred_test

# Kernel Density Estimate plot
plt.figure(figsize=(7,5))
sns.kdeplot(residuals, fill=True)
plt.axvline(0, color='red', linestyle='--')
plt.title("Residuals KDE (Test Set)")
plt.xlabel("Residual")
plt.ylabel("Density")
plt.show()





#ANSWER 3.6b: Yes, the residuals look roughly bell-shaped around zero. The KDE plot shows a unimodal, symmetrical distribution centered close to zero (bell-shaped and therefore desirable and thus the residuals are of a well-performing regression model).
#Strengths include that the residuals are centered around zero, so the model's predictions are, on average, unbiased. The bell-shaped distribution tells us that the model's errors follow a normal distribution which is a common assumption used in statistical models. Finally, the high density peak near zero tells us that most of the predictions are close to the actual values. This all tells us that the joint model handles the variables over dummy encoding effectively and since residuals are centered around zero, there is no under/overprediction.
#Weaknesses include that the spread of the residuals is pretty large which means that while the average error may be low, the individual predictions can have pretty big absoulute errors, which could be problematic (depends on context and target var units). So some price ranges are not well predicted. Also, linear models may not capture nonlinear interactions (here, polynomial expansion could help but there is a concern of overfitting). The model also assumes a constant variance of residuals which means that if residual variance increases with price, then predictions for expensive cars could be less and less accurate. 


#QUESTION 4

import sympy as sp
from IPython.display import display, Markdown

sp.init_printing()

# PART 0: Define symbols

N = sp.symbols('N', integer=True, positive=True)
i = sp.symbols('i', integer=True)

b0, b1, b2 = sp.symbols('b_0 b_1 b_2')
ybar = sp.symbols(r'\bar{y}')

y = sp.IndexedBase('y')
z1 = sp.IndexedBase('z_1')
z2 = sp.IndexedBase('z_2')

display(Markdown("## Part 0: Model"))
yhat_i = b0 + b1*z1[i] + b2*z2[i]
display(sp.Eq(sp.Symbol(r'\hat{y}_i'), yhat_i))

# PART 1: Write down the SSE
display(Markdown("## Part 1: Sum of Squared Errors (SSE)"))

SSE = sp.summation((y[i] - (b0 + b1*z1[i] + b2*z2[i]))**2, (i, 1, N))
display(sp.Eq(sp.Symbol('SSE'), SSE))

# PART 2: Take partial derivatives with respect to b0, b1, b2

display(Markdown("## Part 2: Partial Derivatives"))

dSSE_db0 = sp.diff(SSE, b0)
dSSE_db1 = sp.diff(SSE, b1)
dSSE_db2 = sp.diff(SSE, b2)

display(sp.Eq(sp.Symbol(r'\frac{\partial SSE}{\partial b_0}'), dSSE_db0))
display(sp.Eq(sp.Symbol(r'\frac{\partial SSE}{\partial b_1}'), dSSE_db1))
display(sp.Eq(sp.Symbol(r'\frac{\partial SSE}{\partial b_2}'), dSSE_db2))

# PART 3: Set derivatives equal to zero and verify conditions
# average error = 0 and e·z = 0 at optimum

display(Markdown("## Part 3: First-Order Conditions"))

# Define residual/error
e = y[i] - (b0 + b1*z1[i] + b2*z2[i])

eq_b0 = sp.Eq(sp.summation(e, (i, 1, N)), 0)
eq_b1 = sp.Eq(sp.summation(z1[i]*e, (i, 1, N)), 0)
eq_b2 = sp.Eq(sp.summation(z2[i]*e, (i, 1, N)), 0)

display(Markdown("From setting the derivatives equal to zero:"))
display(eq_b0)
display(eq_b1)
display(eq_b2)

display(Markdown("These mean:"))
display(Markdown(r"- The average residual is zero: $\frac{1}{N}\sum_{i=1}^N e_i = 0$"))
display(Markdown(r"- Residuals are orthogonal to $z_1$: $\sum_{i=1}^N z_{i1} e_i = 0$"))
display(Markdown(r"- Residuals are orthogonal to $z_2$: $\sum_{i=1}^N z_{i2} e_i = 0$"))

# PART 4: Show that optimal intercept is b0* = ybar
# Then eliminate b0 and focus on b1, b2

display(Markdown("## Part 4: Solve for the Intercept"))

display(Markdown(
    r"Using $\sum e_i = 0$: "
    r"$\sum_{i=1}^N (y_i - b_0 - b_1 z_{i1} - b_2 z_{i2}) = 0$"
))

display(Markdown(
    r"Because the predictors are centered, "
    r"$\sum_{i=1}^N z_{i1}=0$ and $\sum_{i=1}^N z_{i2}=0$."
))

b0_star = ybar
display(sp.Eq(sp.Symbol(r'b_0^*'), b0_star))

display(Markdown(
    r"So after substituting $b_0^*=\bar{y}$, the remaining equations only involve $b_1$ and $b_2$."
))

eq1_reduced = sp.Eq(
    sp.summation(z1[i]*(y[i] - ybar - b1*z1[i] - b2*z2[i]), (i, 1, N)),
    0
)

eq2_reduced = sp.Eq(
    sp.summation(z2[i]*(y[i] - ybar - b1*z1[i] - b2*z2[i]), (i, 1, N)),
    0
)

display(eq1_reduced)
display(eq2_reduced)

# PART 5: Write results as matrix equation Ab = C

display(Markdown("## Part 5: Matrix Form $Ab = C$"))

S11 = sp.Symbol(r'\sum z_{i1}^2')
S22 = sp.Symbol(r'\sum z_{i2}^2')
S12 = sp.Symbol(r'\sum z_{i1}z_{i2}')
C1  = sp.Symbol(r'\sum z_{i1}(y_i-\bar{y})')
C2  = sp.Symbol(r'\sum z_{i2}(y_i-\bar{y})')

A = sp.Matrix([
    [S11, S12],
    [S12, S22]
])

b_vec = sp.Matrix([
    [b1],
    [b2]
])

C = sp.Matrix([
    [C1],
    [C2]
])

display(Markdown("Matrix $A$:"))
display(A)

display(Markdown("Vector $b$:"))
display(b_vec)

display(Markdown("Vector $C$:"))
display(C)

display(Markdown("Normal equations:"))
display(sp.Eq(A*b_vec, C))

# PART 6: Divide by N and substitute z_ij = x_ij - m_j
# Identify A and C and explain intuition

display(Markdown("## Part 6: Divide by $N$ and Interpret"))

m1, m2 = sp.symbols('m_1 m_2')
x1 = sp.IndexedBase('x_1')
x2 = sp.IndexedBase('x_2')

A_avg = sp.Matrix([
    [sp.Symbol(r'\frac{1}{N}\sum z_{i1}^2'), sp.Symbol(r'\frac{1}{N}\sum z_{i1}z_{i2}')],
    [sp.Symbol(r'\frac{1}{N}\sum z_{i1}z_{i2}'), sp.Symbol(r'\frac{1}{N}\sum z_{i2}^2')]
])

C_avg = sp.Matrix([
    [sp.Symbol(r'\frac{1}{N}\sum z_{i1}(y_i-\bar{y})')],
    [sp.Symbol(r'\frac{1}{N}\sum z_{i2}(y_i-\bar{y})')]
])

display(Markdown("After dividing both sides by $N$:"))
display(sp.Eq(A_avg * b_vec, C_avg))

display(Markdown(
    r"Now use $z_{i1}=x_{i1}-m_1$ and $z_{i2}=x_{i2}-m_2$."
))

display(Markdown("So:"))
display(Markdown(
    r"- $A$ is the covariance matrix of the centered predictors"
))
display(Markdown(
    r"- $C$ is the covariance vector between each centered predictor and the centered response"
))

display(Markdown(
    r"In intuition: the regression coefficients solve a system that uses "
    r"how the predictors vary with each other and how each predictor varies with $y$."
))
```
